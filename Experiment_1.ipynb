```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
#1000 random integers between 0 and 50
x=np.random.randint(0, 50, 1000)
print(x)
```

```python
#positive correlation with noise
y= x + np.random.normal(0, 10, 1000)
```

```python
np.corrcoef(x, y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
plt.scatter(x, y)
plt.show()
```

```python
#1000 random integers between 0 and 50
x=np.random.randint(0, 50, 1000)
#negative correlation with some noise
y= 100 - x + np.random.normal(0, 10, 1000)
```

```python
np.corrcoef(x, y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
plt.scatter(x, y)
plt.show()
```
